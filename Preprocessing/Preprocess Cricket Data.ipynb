## Batting Partnership Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Input/IPL Ball-by-Ball 2008-2020.csv")
df
```

```python
df2 = df.groupby(['id','batting_team','batsman','non_striker']).agg({'total_runs': ['sum'],'ball': ['count']})
df2.columns = ['runs', 'balls']
df2 = df2.reset_index()
```

```python
df2.batting_team.unique()
```

```python
df_team = df2[df2['batting_team'] == 'Chennai Super Kings'].copy()
```

```python
df_team = df_team.reset_index(drop = True)
```

```python
df_team
```

```python
newdf = pd.DataFrame(columns = ['id','b1','b2','run','ball']) 
for index, row in df_team.iterrows():
#     print(row['batsman'], row['non_striker'])
    a1 = df_team[df_team['id']==row['id']][df_team['batsman']==row['batsman']][df_team['non_striker']==row['non_striker']]['runs'].values
    a2 = df_team[df_team['id']==row['id']][df_team['non_striker']==row['batsman']][df_team['batsman']==row['non_striker']]['runs'].values
    if a2.size < 1:
        a2 = [0]
    b1 = df_team[df_team['id']==row['id']][df_team['batsman']==row['batsman']][df_team['non_striker']==row['non_striker']]['balls'].values
    b2 = df_team[df_team['id']==row['id']][df_team['non_striker']==row['batsman']][df_team['batsman']==row['non_striker']]['balls'].values
    if b2.size < 1:
        b2 = [0]
#     print(a1 + a2, b1 + b2)
    c1 = newdf[newdf['b1']==row['batsman']][newdf['b2']==row['non_striker']][newdf['id'] == row['id']].values
    c2 = newdf[newdf['b1']==row['non_striker']][newdf['b2']==row['batsman']][newdf['id'] == row['id']].values
    if c1.size > 0 or c2.size > 0:
        continue
    newdf = newdf.append({'id': int(row['id']),'b1': row['batsman'],'b2': row['non_striker'],'run': (a1+a2)[0], 'ball': (b1+b2)[0]}, ignore_index = True)
```

```python
newdf['str'] = (newdf['run']/newdf['ball'])*100
```

```python
import numpy as np
newdf_grouped = newdf.groupby(['b1','b2']).agg({'str': ['sum','count']})
newdf_grouped.columns = ['totalstr', 'countstr']
newdf_grouped = newdf_grouped.reset_index()


newdf2 = pd.DataFrame(columns = ['b1','b2','totalstr','countstr']) 
for index, row in newdf_grouped.iterrows():
    t1 = newdf_grouped[newdf_grouped['b1']==row['b1']][newdf_grouped['b2']==row['b2']]['totalstr'].values
    c1 = newdf_grouped[newdf_grouped['b1']==row['b1']][newdf_grouped['b2']==row['b2']]['countstr'].values
    t2 = newdf_grouped[newdf_grouped['b1']==row['b2']][newdf_grouped['b2']==row['b1']]['totalstr'].values
    c2 = newdf_grouped[newdf_grouped['b1']==row['b2']][newdf_grouped['b2']==row['b1']]['countstr'].values
    if t2.size < 1:
        t2 = [0]
        c2 = [0]
    
    d1 = newdf2[newdf2['b1']==row['b1']][newdf2['b2']==row['b2']].values
    d2 = newdf2[newdf2['b1']==row['b2']][newdf2['b2']==row['b1']].values
    
    if d1.size > 0 or d2.size > 0:
        continue
    newdf2 = newdf2.append({'b1': row['b1'],'b2': row['b2'],'totalstr': (t1+t2)[0], 'countstr': (c1+c2)[0]}, ignore_index = True)
        
newdf2['avgstr'] = newdf2['totalstr']/newdf2['countstr'] 
newdf2
df_team = newdf2.copy()
```

```python
df_team['normstr'] = (((df_team['avgstr']-df_team['avgstr'].min())/(df_team['avgstr'].max()-df_team['avgstr'].min()))*90)+10
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../Input/IPL Ball-by-Ball 2008-2020.csv")

df2 = df.groupby(['id','bowling_team','bowler']).agg({'total_runs': ['sum'],'ball': ['count']})
df2.columns = ['runs', 'balls']
df2 = df2.reset_index()

for team_name in df2.bowling_team.unique():
    df_team = df2[df2['bowling_team'] == team_name].copy()
    df_team = df_team.reset_index(drop = True)
    
    df_team['runrate'] = (df_team['runs']/df_team['balls'])*6
    
    newdf_grouped = df_team.groupby(['bowler']).agg({'runrate': ['sum','count']})
    newdf_grouped.columns = ['totalrr', 'countrr']
    newdf_grouped = newdf_grouped.reset_index()
    
    newdf_grouped['avgrr'] = newdf_grouped['totalrr']/newdf_grouped['countrr']
    newdf_grouped['b1'] = newdf_grouped['bowler']
    
    newdf_grouped['normrr'] = (((newdf_grouped['avgrr']-newdf_grouped['avgrr'].min())/(newdf_grouped['avgrr'].max()-newdf_grouped['avgrr'].min()))*90)+10
    newdf_grouped['normrr'] = (100-(newdf_grouped['normrr']-10))
    
    final_team = newdf_grouped.drop(columns = ['totalrr','countrr'])
    final_team.rename(columns={'bowler': 'Source'},inplace = True)
    final_team.to_csv('Files/Bowling TeamWise/'+team_name+'_bowling_nodelist.csv', index=False)
```

## Individual Batsman Performance

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../Input/IPL Ball-by-Ball 2008-2020.csv")

df2 = df.groupby(['id','batting_team','batsman']).agg({'total_runs': ['sum'],'ball': ['count']})
df2.columns = ['runs', 'balls']
df2 = df2.reset_index()

for team_name in df2.batting_team.unique():
    df_team = df2[df2['batting_team'] == team_name].copy()
    df_team = df_team.reset_index(drop = True)
    
    df_team['str'] = (df_team['runs']/df_team['balls'])*100
    
    newdf_grouped = df_team.groupby(['batsman']).agg({'str': ['sum','count']})
    newdf_grouped.columns = ['totalstr', 'countstr']
    newdf_grouped = newdf_grouped.reset_index()
    
    newdf_grouped['avgstr'] = newdf_grouped['totalstr']/newdf_grouped['countstr']
    
    newdf_grouped['normstr'] = (((newdf_grouped['avgstr']-newdf_grouped['avgstr'].min())/(newdf_grouped['avgstr'].max()-newdf_grouped['avgstr'].min()))*90)+10
    
    final_team = newdf_grouped.drop(columns = ['totalstr','countstr'])
    final_team.rename(columns={'batsman': 'Source'},inplace = True)
    final_team.to_csv('Files/Batting TeamWise/'+team_name+'_batting_nodelist.csv', index=False)
```

## Overall Batting Performance

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../Input/IPL Ball-by-Ball 2008-2020.csv")

df2 = df.groupby(['id','batsman']).agg({'total_runs': ['sum'],'ball': ['count']})
df2.columns = ['runs', 'balls']
df2 = df2.reset_index()

df2['str'] = (df2['runs']/df2['balls'])*100

newdf_grouped = df2.groupby(['batsman']).agg({'str': ['sum','count']})
newdf_grouped.columns = ['totalstr', 'countstr']
newdf_grouped = newdf_grouped.reset_index()

newdf_grouped['avgstr'] = newdf_grouped['totalstr']/newdf_grouped['countstr']
    
newdf_grouped['normstr'] = (((newdf_grouped['avgstr']-newdf_grouped['avgstr'].min())/(newdf_grouped['avgstr'].max()-newdf_grouped['avgstr'].min()))*90)+10

final_team = newdf_grouped.drop(columns = ['totalstr','countstr'])
final_team.rename(columns={'batsman': 'Source'},inplace = True)
final_team.to_csv('Files/Batting Overall/Overall_batting_nodelist.csv', index=False)
```
